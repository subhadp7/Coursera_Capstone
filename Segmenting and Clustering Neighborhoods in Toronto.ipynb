### 2. Web Scraping the Wiki page to get data

```python
# request html link
source=requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup= BeautifulSoup(source,'lxml')
print(soup.title)
```

```python
tab = str(soup.table)
display_html(tab,raw=True)
```

### 3. Convert into pandas dataframe

```python
df1=pd.read_html(tab)
df=df1[0]
df
```

```python
print(df.shape)
```

Now we remove rows from that data fram that doesn't have assigned Borough or Neighborhood

```python
dfa=df[df.Borough != 'Not assigned']
print(dfa.shape)
```

```python
dfa
```

Combining Neighborhoods with same postel codes

```python
dfb=dfa.groupby(['Postal Code','Borough'],sort=False).agg(','.join)
dfb.reset_index(inplace=True)
dfb
```

Replacing the name of the neighbourhoods which are 'Not assigned' with names of Borough

dfb['Neighbourhood']= np.where(dfb['Neighbourhood']=='Not assigned',dfb['Borough'],dfb['Neighbourhood'])
dfb

Final shape of the dataframe

```python
print(dfb.shape)
```

### 4. Adding Latitudes and Longitudes of each Neighborhood in the dataframe

We load the CSV file containing all the latitude and longitude for Neighborhoods

```python
cor= pd.read_csv('https://cocl.us/Geospatial_data')
cor
```

Now merge the two tables on basis of the Postal Codes

```python
dfc=pd.merge(dfb,cor,on='Postal Code')
```

```python
dfc
```

### 5.Exploring ,clustering and visualizing the neighbourhoods in Toronto
#### (Only working with boroughs with word 'Toronto' in it)

```python
toronto_df=dfc[dfc['Borough'].str.contains('Toronto',regex=False)]
toronto_df
```

Using folium to show all the points in map

```python
toronto_latitude= 43.6532
toronto_longitude= -79.3832

toronto_map= folium.Map(location=[toronto_latitude,toronto_longitude],zoom_start=10)
for p,b,n,la,lo in zip(toronto_df['Postal Code'],toronto_df['Borough'],toronto_df['Neighbourhood'],toronto_df['Latitude'],toronto_df['Longitude']):
    label= '{},{}'.format(n,b)
    label= folium.Popup(label,parse_html=True)
    folium.CircleMarker(
    [la,lo],
    radius=5,
    color='blue',
    fill_color='#3186cc',
    fill_opacity=0.7,
    popup=label,
    parse_html = False).add_to(toronto_map)

toronto_map    
```

Now we apply KMeans clustering 

```python
toronto_cor= toronto_df.drop(['Postal Code','Borough','Neighbourhood'],1)
kmeans=KMeans(n_clusters=5,random_state=0).fit(toronto_cor)
kmeans.labels_
```

```python
toronto_df.insert(0,'Cluster',kmeans.labels_)
toronto_df
```

Creating Map with Clusters

```python
map_clusters= folium.Map(location=[toronto_latitude,toronto_longitude],zoom_start=10)

x = np.arange(5)
ys = [i + x + (i*x)**2 for i in range(5)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

markers_colors = []
for la, lo, n, c in zip(toronto_df['Latitude'], toronto_df['Longitude'], toronto_df['Neighbourhood'], toronto_df['Cluster']):
    label = folium.Popup(' Cluster ' + str(c), parse_html=True)
    folium.CircleMarker(
        [la, lo],
        radius=5,
        popup=label,
        color=rainbow[c-1],
        fill=True,
        fill_color=rainbow[c-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
